```python
from datetime import datetime    #زمان
import pandas as pd                    #پانداس
import matplotlib.pyplot as plt         #مصورسازی
import numpy as np          
import finpy_tse as fpy            # دریافت دیتای قیمتی از finpy
import pytse_client as tse
import yfinance as yf
import ta 
import math as math
import asyncio
```

```python


# دریافت داده قیمتی شاخص کل بورس

def index_kol(bars:int=9500):
    while True:
        try:
            data=fpy.Get_CWI_History(ignore_date=True,double_date=True)
            data=data.iloc[:-bars:-1][::-1]
            data.index=data['Date']
            return(data)
        except Exception as e:
            print(f"An error occurred: {e}")
            asyncio.sleep(2)  # Sleep for 10 seconds before retrying



symbol_data=index_kol()
symbol_data
```

```python
symbol_data['ma49']=symbol_data['Close'].rolling(49).mean()    #محاسبه مووینگ 49 روزه ساده 
```

```python
len_symbol_data=len(symbol_data)
len_symbol_data
```

```python
def trend_status(shiftex:int=2):

    symbol_data['uptrend']=np.zeros((len_symbol_data))
    symbol_data['downtrend']=np.zeros((len_symbol_data))
    symbol_data['confused']=np.zeros((len_symbol_data))


    for i in range (-1,-(len_symbol_data)+2,-1):
        
        if symbol_data['Close'][i+shiftex] > symbol_data['ma49'][i+shiftex]   and   symbol_data['ma49'][i+shiftex] > symbol_data['ma49'][i+shiftex-1] :
            
            symbol_data['uptrend'][i+shiftex]= 'up'
        
        
        elif symbol_data['Close'][i+shiftex] < symbol_data['ma49'][i+shiftex]   and   symbol_data['ma49'][i+shiftex] < symbol_data['ma49'][i+shiftex-1] :
            
            symbol_data['downtrend'][i+shiftex]= 'down'

        else:
            symbol_data['confused'][i+shiftex]= 'confused'
            
        

    return symbol_data




trend_status()
```
